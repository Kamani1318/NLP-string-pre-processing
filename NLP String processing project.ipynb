```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
def extract_article_text(url):
    # Make a request to the website
    response = requests.get(url)

    # Create BeautifulSoup object and parse the HTML content
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find the HTML elements containing the article text
    article_elements = soup.find_all('p')  # Adjust this based on the structure of the web page

    # Extract the text from the article elements
    article_text = ' '.join(element.get_text() for element in article_elements)

    return article_text
```

# Positive and Negative Score

```python
positive_score = []
negative_score = []
total_words_per_parah = []
total_words = []
characters_per_parah = []
for i in range(filtered_text_df.shape[0]):
    words1 = nltk.word_tokenize(filtered_text_df.iloc[i,0])
    words1 = [re.sub(r'[^\w\s]', '', word) for word in words1]
    words1 = [word for word in words1 if word != '']
    characters = sum(len(word) for word in words1)
    characters_per_parah.append(characters)
    total_words_per_parah.append(len(words1))
    total_words.extend(words1)
    matching_strings1 = set(words1) & set(positive_words)
    matching_strings2 = set(words1) & set(negative_words)
    positive_score.append(len(matching_strings1))
    negative_score.append(len(matching_strings2))
```

```python
target_words = ["i", "we", "my", "ours", "us"]

# Tokenize the paragraph into words
personal_pronouns = []

for i in range(article_text_df.shape[0]):
    words = re.findall(r'\w+',article_text_df.iloc[i,0])
    words = [word.lower() for word in words]
    word_counts = {word: 0 for word in target_words}

    for word in words:
        if word in target_words:
            word_counts[word] += 1
    personal_pronouns.append(sum(word_counts.values()))
```

# Syllable Count Per Word

```python
def remove_suffixes(word):
    pattern = r"(es|ed)$"
    result = re.sub(pattern, "", word)
    return result
```

```python
vowels = ['a','e','i','o','u']
syllable_count_per_word_list = []
for i in range(filtered_text_df.shape[0]):
    words1 = nltk.word_tokenize(filtered_text_df.iloc[i,0])
    words1 = [re.sub(r'[^\w\s]', '', word) for word in words1]
    words1 = [word for word in words1 if word != '']
    syllable_count_of_parah = 0
    for word in tqdm(words1):
        modified_word = remove_suffixes(word)
        # characters_of_word = set(word)
        syllable_count = len(set(vowels) & set(modified_word))
        syllable_count_of_parah = syllable_count_of_parah + syllable_count
    syllable_count_per_word_list.append(syllable_count_of_parah/len(words1))
```
